```python
#Se eliminan columnas diferentes a estrato 1, 2, 3, 4, 5, 6 , se consideran como error, la cantidad de datos que se eliminan son muy pequeña en comparación
Validos=["Estrato 1", "Estrato 2","Estrato 3","Estrato 4","Estrato 5","Estrato 6"]
df = df[df["fami_estratovivienda"].isin(Validos)]
#Se verifica otra vez el numero 
df["fami_estratovivienda"].value_counts() 
```

```python
#Ver cuantos datos tienen esta particularidad
df[df["punt_matematicas"] > 100].shape[0]
```

```python
#Ver las filas
df[df["punt_matematicas"] > 100]
```

```python
df = df[~(
    (df["punt_ingles"] > 100) |
    (df["punt_matematicas"] > 100) |
    (df["punt_sociales_ciudadanas"] > 100) |
    (df["punt_c_naturales"] > 100) |
    (df["punt_lectura_critica"] > 100)
)]
```

```python
df.isnull().sum()
```

```python
#Muchos Null en la columna cole_bilingue, se procede a crear una categoria nueva.
df["cole_bilingue"] = df["cole_bilingue"].fillna("No reportado")
```

```python
df.isnull().sum()
```

```python
#Categorias de la categoria cole_caracter
df["cole_caracter"].unique()
```

```python
#Se hace lo mismo con la columna cole_caracter
df["cole_caracter"] = df["cole_caracter"].fillna("NO REPORTADO")
```

```python
df.isnull().sum()
```

```python
#Se observa que el numero de nulls en algunas pruebas especificas del examen es el mismo y que las pruebas de matematica e ingles son 0,
#  eso puede significar algo, entonces por ahora se dejan esas y se eliminan el resto porque en total son 210 filas no es ni el 0.5% de la muestra.
columnas_con_NA= [
    "cole_cod_dane_establecimiento",
    "estu_cod_reside_depto",
    "estu_cod_reside_mcpio",
    "estu_depto_reside",
    "estu_genero",
    "estu_mcpio_reside"
]

df = df.dropna(subset=columnas_con_NA)
```

```python
df.isnull().sum()
```

```python
df.info()
#quedaron 566796 filas al final.
```

```python
#Crear una nueva columna para decir si presentaron todas las areas o no.
df["presento_todas_las_areas"] = df["punt_global"].notna()
```
