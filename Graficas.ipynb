```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Icfes_limpio.csv')
```

```python
df.info()
```

```python
#Distinguir variables.
# Ver tipos de datos
print(df.dtypes)

# Separar columnas numéricas y categóricas
numericas = df.select_dtypes(include="number").columns.tolist()
categoricas = df.select_dtypes(include="object").columns.tolist()

print("Numéricas:", numericas)
print("Categóricas:", categoricas)
```

```python
df["periodo"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.histplot(df["periodo"], kde=False)
plt.title(f"Distribución de peroido")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
interes=["punt_ingles","punt_matematicas", "punt_sociales_ciudadanas","punt_c_naturales","punt_lectura_critica","punt_global"]
# Histograma para columnas numéricas
for col in interes:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribución de {col}")
    plt.show()

```

```python
df['cole_area_ubicacion'].value_counts()
```

```python

# Gráfico de barras para categóricas

interesc=['estu_tipodocumento','cole_area_ubicacion', 'cole_bilingue','cole_calendario', 'cole_caracter', 'cole_genero', 'cole_jornada', 'estu_genero', 'estu_nacionalidad', 'estu_pais_reside', 'estu_privado_libertad', 'fami_cuartoshogar', 'fami_educacionmadre', 'fami_educacionpadre', 'fami_estratovivienda', 'fami_personashogar', 'fami_tieneautomovil', 'fami_tienecomputador', 'fami_tieneinternet', 'fami_tienelavadora', 'desemp_ingles']
for col in interesc:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Frecuencia de {col}")
    plt.show()
```

```python
# Ver valores faltantes
print(df.isnull().sum())
```

```python
correlacion = df.corr(numeric_only=True)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

```python
columnas_con_nan = ["punt_sociales_ciudadanas", "punt_c_naturales", "punt_lectura_critica", "punt_global"]

# Crear el mapa de calor
plt.figure(figsize=(10, 4))
sns.heatmap(df[columnas_con_nan].isnull(), cbar=False, yticklabels=False)
plt.title("Valores faltantes en puntajes por área")
plt.show()
```

```python
import seaborn as sns

df["tiene_nan"] = df["punt_global"].isnull()

sns.countplot(data=df, x="periodo", hue="tiene_nan")
plt.title("Cantidad de estudiantes por periodo y presencia de NaN en puntaje global")
plt.show()
```
